#### Step 9: Data Visualization

1. Let's start with analyzing the Electric Vehicles Adoption Over Time by visualizing the number of Electric Vehicles registered by model year. It will give us an insight into how the Electric Vehicles has grown over the years.

```python
# Electric Vehicles Adoption Over Time
plt.figure(figsize=(12, 6))
ev_adoption_by_year = df['Model Year'].value_counts().sort_index()
sns.barplot(x=ev_adoption_by_year.index, y=ev_adoption_by_year.values, palette="viridis")

plt.title('Electric Vehicles Adoption Over Time')
plt.xlabel('Model Year')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

From the above bar chart, it’s clear that Electric Vehicles adoption has been increasing over time, especially noting a significant upward trend starting around 2016. The number of vehicles registered grows modestly up until that point and then begins to rise more rapidly from 2017 onwards. The year 2023 shows a particularly sharp increase in the number of registered Electric Vehicles, with the bar for 2023 being the highest on the graph, indicating a peak in Electric Vehicles adoption

2. Next, let’s explore the types of electric vehicles represented in this dataset. So, let’s visualize the distribution of electric vehicles types to see which categories are most popular among the registered vehicles:

```python
# Analyzing the distribution of electric vehicles Types
ev_type_distribution = df['Electric Vehicle Type'].value_counts()

plt.figure(figsize=(10, 6))
plt.pie(ev_type_distribution, labels=ev_type_distribution.index, autopct='%1.1f%%', startangle=65)
plt.title('Distribution of Electric Vehicles Types')
plt.show()
```

The above Pie chart shows that BEVs are more popular or preferred over PHEVs among the electric vehicles.

3. Let's now focus on the popularity of electric vehicles manufacturers and models among the registered vehicles. This analysis will help us identify which manufacturers and specific models dominate the Electric Vehicles market, potentially indicating consumer preferences, brand loyalty, and the success of various manufacturers’ strategies in promoting electric mobility.

   So, let’s have a look at the most popular manufacturers and then drill down into the most popular models within those manufacturers:

```python
# Analyzing the popularity of EV manufacturers
ev_make_distribution = df['Make'].value_counts().head(10)  # Limiting to top 10 for clarity

plt.figure(figsize=(12, 6))
sns.barplot(x=ev_make_distribution.values, y=ev_make_distribution.index, palette="cubehelix")
plt.title('Top 10 Popular Electric Vehicles Makes')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('Make')
plt.show()
```

The above chart shows that:

-TESLA leads by a substantial margin with the highest number of vehicles registered.

-NISSAN is the second most popular manufacturer, followed by CHEVROLET, though both have significantly fewer registrations than TESLA.

-FORD, BMW, KIA, TOYOTA, VOLKSWAGEN, VOLVO, and JEEP follow in decreasing order of the number of registered vehicles.

##### 4. Bar Chart

```python
# To view the top 5 vs bottom 5 cars comparision
car_market_share = pd.DataFrame(df.groupby('Make')['VIN (1-10)'].count().reset_index()).sort_values(by='VIN (1-10)',ascending=False).reset_index(drop=True)
car_market_share.columns = ['model','count']
car_market_share_top_5 = car_market_share.head(5)
car_market_share_bottom_5 = car_market_share.tail(5)

fig, axes = plt.subplots(1, 2, figsize=(19, 5))
sns.barplot(data=car_market_share_top_5, x="model", y="count",ax=axes[0], palette="cubehelix")
sns.barplot(data=car_market_share_bottom_5, x="model", y="count",ax=axes[1], palette="cubehelix")
axes[0].set_title('Top 5 Cars')
axes[1].set_title('Bottom 5 Cars')
```

TESLA is the market leader by large.

Top 5 cars are TESLA,NISSAN,CHEVROLET model,FORD,BMW.

Bottom cars are FISKER,AZURE DYNAMICS,THINK model,BENTLEY,WHEEGO ELECTRIC CARS

##### 5. Line Chart

```python
# Calculating the average electric range by model year
average_range_by_year = df.groupby('Model Year')['Electric Range'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='Model Year', y='Electric Range', data=average_range_by_year, marker='o', color='green')
plt.title('Average Electric Range by Model Year')
plt.xlabel('Model Year')
plt.ylabel('Average Electric Range (miles)')
plt.show()
```

-Trends Over Time: The x-axis represents the model years, while the y-axis shows the average electric range in miles, highlighting how the range has changed over the years.

-Data Points: Each point on the line is marked, showing the average range for that specific year, providing a clear visual of trends.

-Key Insight: This visualization reveals how advancements in technology and design have influenced electric vehicle performance, allowing viewers to   identify trends and improvements in electric range over time.

##### 6. Bar Chart 

```python
average_range_by_model = df.groupby(['Make', 'Model'])['Electric Range'].mean().sort_values(ascending=False).reset_index()

# The top 10 models with the highest average electric range
top_range_models = average_range_by_model.head(10)

plt.figure(figsize=(12, 7))
barplot = sns.barplot(x='Electric Range', y='Model', hue='Make', data=top_range_models, palette="cool")
plt.title('Top 10 Models by Average Electric Range in Top Makes')
plt.xlabel('Average Electric Range (miles)')
plt.ylabel('Model')
plt.legend(title='Make', loc='center right')
plt.show()
```

-Models Highlighted: It showcases the models that offer the greatest distance on a full charge, providing insights into their performance.

-Average Electric Range: The x-axis shows the average range in miles, allowing for easy comparison of how far each model can go.

-Manufacturer Differentiation: Different colors represent various manufacturers (Make), helping to visualize which brands offer the highest-performing models.

```python
# 3. Distribution based on County
county_counts = df['County'].value_counts().head(7)
plt.figure(figsize=(11, 7))
plt.pie(county_counts, labels=county_counts.index, autopct='%1.1f%%', startangle=55)
plt.title('Distribution of Vehicles by Country')
plt.axis('equal')
plt.show()
```
